# Phase 2: Exploratory Visuals

Step 1: Import Visualization Libraries

i now import visualization libraries that will help me explore patterns in the weather data.


matplotlib and Seaborn allow me to create charts that make it easier to understand trends and relationships between variables.

```python
# importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Actual Visualiasations

Visualization 1: Distribution of Maximum Daily Temperature

```python
# creating a histogram to see how maximum temperatures are distributed
plt.figure()

sns.histplot(df['TMAX'], bins=50)

# adding title and axis labels
plt.title("Distribution of Maximum Daily Temperature (2017)")
plt.xlabel("Maximum Temperature")
plt.ylabel("Frequency")

plt.show()
```



*Visialisation 1 graph explained: x-axis have the maximum temperature recorded while y-axis show the frequency or how many times the temperature value appers in the data, therefore the graph shows high frequency between about 40 and 100 which means that most the days recorded maximum temperatures within that range. very low and very high tempretures dont appear that much, which is why the bars on the edges of the histogram are smaller*




Visualization 2: Distribution of Minimum Daily Temperature

```python
# histogram showing the distribution of minimum temperatures
plt.figure()

sns.histplot(df['TMIN'], bins=50)

plt.title("Distribution of Minimum Daily Temperature (2017)")
plt.xlabel("Minimum Temperature")
plt.ylabel("Frequency")

plt.show()
```

*Visualisation 2 graph explained: here, rhe x-axis shows the minimum temperature and the y axis the frequency. most of these values cluster in moderate middle range between 35 anf 50 degrees. the smaller bars on the left show that very cold nights are not that common compared to typical cool nights*

Visualization 3: Boxplot of Maximum Temperature

```python
# creating a boxplot to detect outliers in maximum temperatures
plt.figure()

sns.boxplot(x=df['TMAX'])

# adding title
plt.title("Boxplot of Maximum Daily Temperature (2017)")

plt.xlabel("Maximum Temperature")

plt.show()
```

*Visualisation 3 graph explained: in this boxplot the x-axis represent the maximum temperature recorded in the dataset. that box in the middle shows where the bulk of the data falls, between 40 and 85. moving on, the line inside the box represents the median which is around 66-70. the points way out past the lines are outliers, some of these reach as high as 261.68, which is a huge red flag because that temperature is impossible for normal weather. it shows us exactly where there are errors in our data file.*

Visualization 4: Distribution of Daily Precipitation

```python
# histogram showing distribution of precipitation
plt.figure()

sns.histplot(df['PRCP'], bins=50)

plt.title("Distribution of Daily Precipitation (2017)")
plt.xlabel("Precipitation")
plt.ylabel("Frequency")

plt.show()
```

*Visualisation 4 graph explained: x-axis shows the amount of rain /precipitation and the y-axis shows the frequency. most data is shown toward the 0.00 rain which i could say it represents dry rains, the average rainfall is is 0.11, the bars in the grapgh drop off almost instantly*

Visualization 5: Distribution of Snowfall

```python
# histogram for snowfall values
plt.figure()

sns.histplot(df['SNOW'].dropna(), bins=50)

plt.title("Distribution of Snowfall")
plt.xlabel("Snowfall")
plt.ylabel("Frequency")

plt.show()
```

*Visualisation 5 graph explained: in this bar grapgh the x axis shows the snowfall amounts and the yaxis is the frequency. same as the rain, almost all the data is at 0, which means most stations had no snow, indicating that snow events occur less frequently in the dataset.*

Visualization 6: Average Maximum Temperature by Month

```python
# extracting month from date
df['month'] = df['date'].dt.month

# calculating average monthly max temperature
monthly_temp = df.groupby('month')['TMAX'].mean()

plt.figure()

monthly_temp.plot(kind='line')

plt.title("Average Maximum Temperature by Month (2017)")
plt.xlabel("Month")
plt.ylabel("Temperature")

plt.show()
```

*Visualisation 6 graph explained: i now have a line chart and it just shows the average maximum temperature for each month in 2017. we can see the temperatures gradually increase during the middle months of the year and decrease during the earlier and later months, demonstrating a clear seasonal temperature pattern. the chart kind of created a mountain shape and gives a clear look at the seasonal cycle throughout the year*

Visualization 7: Total Precipitation by Month

```python
# calculating monthly precipitation totals
monthly_rain = df.groupby('month')['PRCP'].sum()

plt.figure()

monthly_rain.plot(kind='bar')

plt.title("Total Precipitation by Month (2017)")
plt.xlabel("Month")
plt.ylabel("Total Precipitation")

plt.show()
```

*Visualisation 7 graph explained: i have the x-axis showing the months of the year, and the y-axis the total amount of rain added up for all stations. so, some months show noticeably higher rainfall totals than others, suggesting that precipitation varies throughout the year rather than being evenly distributed. For example, the bars for the first months are higher than the rest of the months*

Visualization 8: Temperature vs Precipitation

```python
# scatter plot comparing temperature and rainfall
plt.figure()

sns.scatterplot(x=df['TMAX'], y=df['PRCP'])

plt.title("Temperature vs Precipitation")
plt.xlabel("Maximum Temperature")
plt.ylabel("Precipitation")

plt.show()
```

*Visualisation 8 scatter plot explained: the scatter plot i have above explores the relationship between temperature and rain, the x axis being the maximum temperature and the y-axis the amount of precipitation. each individual dot represents one day of weather at one station. because the dots are spread out it shows that rain can happen at almost any temperature. however the highest dots which means the heavy storms mostly cluster between 60 and 100 degrees so this shows that the most intense rain usually happens on warmer days.*

Visualization 9: Top 10 States with Most Weather Records

```python
# counting records per state
top_states = df['state'].value_counts().head(10)

plt.figure()

top_states.plot(kind='bar')

plt.title("Top 10 States with Most Weather Records")
plt.xlabel("State")
plt.ylabel("Number of Records")

plt.show()
```

*Visualisation 9 graph explained: this chart basically shows the states with the highest number of weather observations in the dataset. Utah appears to have the most records followed by states such as Montana and California and this easily suggests that the dataset contains a higher concentration of weather stations in these regions.*

Visualization 10: Average Maximum Temperature by State

```python
# calculating average max temperature by state
state_temp = df.groupby('state')['TMAX'].mean().sort_values(ascending=False).head(10)

plt.figure()

state_temp.plot(kind='bar')

plt.title("Top 10 States by Average Maximum Temperature")
plt.xlabel("State")
plt.ylabel("Average Maximum Temperature")

plt.show()
```

*Visualisation 10 graph explained: this visualization compares the average maximum temperatures recorded across different states, x-axis lists the states and the y-axis shows the average maximum temperature for the whole year. that is the taller the bar the hotter the state is on average. this makes it easy to compare the climate of different regions.

For example if Texas has a taller bar than Louisiana, it proves that Texas stays much warmer throughout the year, even though both states are in the same dataset*

Visualization 11: Correlation Heatmap

```python
# selecting only numerical columns
numeric_data = df[['TMAX','TMIN','PRCP','SNOW','SNWD']]

# calculating correlation matrix
correlation = numeric_data.corr()

plt.figure(figsize=(8,6))

sns.heatmap(correlation, annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap of Weather Variables")

plt.show()
```

*Visualisation 11 correlation heatmap explained: the correlation heatmap shows relationships between key weather variables,maximum temperature and minimum temperature show a strong positive correlation, meaning they tend to increase and decrease together. other variables, such as precipitation and snowfall, show weaker relationships with temperature.*

Visualization 12: Daily Temperature Trend Over Time

```python
# sorting by date
df_sorted = df.sort_values('date')

plt.figure()

plt.plot(df_sorted['date'], df_sorted['TMAX'])

plt.title("Maximum Temperature Trend Over Time (2017)")
plt.xlabel("Date")
plt.ylabel("Maximum Temperature")

plt.show()
```

*Visualisation 12 line chart explained: this line chart shows how maximum temperatures change over time throughout 2017. the fluctuations in the line reflect natural weather variability and seasonal changes across the year. the line goes up and down, which represents the daily changes in weather.*

Visualization 13: Average Temperature by Month

```python
# calculating average monthly temperature
monthly_avg_temp = df.groupby('month')['TMAX'].mean()

plt.figure()

monthly_avg_temp.plot(kind='bar')

plt.title("Average Maximum Temperature by Month")
plt.xlabel("Month")
plt.ylabel("Average Temperature")

plt.show()
```

*Visualisation 13 graph explained: this last bar chart highlights the seasonal variation in average maximum temperatures across the months. higher temperatures occur during the middle months of the year, while lower temperatures are observed at the beginning and end of the year*
